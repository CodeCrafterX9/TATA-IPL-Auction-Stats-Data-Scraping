```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url="https://www.iplt20.com/auction/2020"
```

```python
r=requests.get(url)
```

```python
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
table=soup.find("table",class_="ih-td-tab w-100 auction-tbl")
print(table)
```

```python
title=table.find_all("th")
print(title)
```

```python
header=[]
for i in title:
    name=i.text
    header.append(name)
print(header)
```

```python
df=pd.DataFrame(columns=header)
print(df)
```

```python
rows=table.find_all("tr")
print(rows)
```

```python
for i in rows[1:]:
    data=i.find_all("td")
    row=[tr.text for tr in data]
    row[1]=row[1].strip( )
    l=len(df)
    df.loc[l]=row
print(df)
```

```python
df.to_csv("ipl_Auction_stats.csv")
```
